```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate Sample Data (2D for visualization)
num_samples = 300
num_clusters = 3

# Creating random clusters
np.random.seed(42)
X1 = np.random.randn(num_samples // num_clusters, 2) + np.array([2, 2])
X2 = np.random.randn(num_samples // num_clusters, 2) + np.array([-2, -2])
X3 = np.random.randn(num_samples // num_clusters, 2) + np.array([2, -2])

# Combine the clusters
X = np.vstack([X1, X2, X3]).astype(np.float32)

# Step 2: Convert to Tensor
X_tensor = tf.convert_to_tensor(X)

# Step 3: Initialize Cluster Centers (Randomly select k points)
k = num_clusters
indices = np.random.choice(X.shape[0], k, replace=False)
centroids = tf.Variable(tf.gather(X_tensor, indices))

# Step 4: K-Means Algorithm
def assign_clusters(X, centroids):
    distances = tf.norm(X[:, None] - centroids, axis=2)  # Compute Euclidean distance
    return tf.argmin(distances, axis=1)

def update_centroids(X, labels, k):
    new_centroids = []
    for i in range(k):
        cluster_points = tf.boolean_mask(X, labels == i)
        new_centroids.append(tf.reduce_mean(cluster_points, axis=0))
    return tf.stack(new_centroids)

# Step 5: Run the K-Means Algorithm
num_iterations = 10
for i in range(num_iterations):
    labels = assign_clusters(X_tensor, centroids)  # Step 5.1: Assign clusters
    centroids.assign(update_centroids(X_tensor, labels, k))  # Step 5.2: Update centroids

# Step 6: Visualize the Clusters
labels = assign_clusters(X_tensor, centroids).numpy()
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', alpha=0.6)
plt.scatter(centroids.numpy()[:, 0], centroids.numpy()[:, 1], c='red', marker='X', s=200, label="Centroids")
plt.legend()
plt.title("K-Means Clustering using TensorFlow/Keras")
plt.show()
```
